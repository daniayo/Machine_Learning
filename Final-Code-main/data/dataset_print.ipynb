1. mon_standard.pkl > array code



```python
import pickle

USE_SUBLABEL = False
URL_PER_SITE = 10
TOTAL_URLS   = 950

print("Loading datafile")
with open("/content/sample_data/mon_standard.pkl", 'rb') as fi:
    data = pickle.load(fi)

X1 = [] # Array to store instances (timestamps) - 19,000 instances, e.g., [[0.0, 0.5, 3.4, ...], [0.0, 4.5, ...], [0.0, 1.5, ...], ... [... ,45.8]]
X2 = [] # Array to store instances (direction*size) - size information
y = [] # Array to store the site of each instance - 19,000 instances, e.g., [0, 0, 0, 0, 0, 0, ..., 94, 94, 94, 94, 94]

# Differentiate instances and sites, and store them in the respective x and y arrays
# x array (direction*timestamp), y array (site label)
for i in range(TOTAL_URLS):
    if USE_SUBLABEL:
        label = i
    else:
        label = i // URL_PER_SITE
         # Calculate which site's URL the current URL being processed belongs to and set that value as the label.
         #Thus, URLs fetched from the same site are labeled identically.
    for sample in data[i]:
        size_seq = []
        time_seq = []
        for c in sample:
            dr = 1 if c > 0 else -1
            time_seq.append(abs(c))
            size_seq.append(dr * 512)
        X1.append(time_seq)
        X2.append(size_seq)
        y.append(label)
size = len(y)

print(f'Total samples: {size}')

print("X1 (first 5 elements):", X1[:5])
print("X2 (first 5 elements):", X2[:5])
print("y (first 5 elements):", y[:5])


```

```python
import pickle
import numpy as np
from sklearn.model_selection import train_test_split

# Load data
with open("/content/sample_data/mon_standard.pkl", 'rb') as f:
    monitored_data = pickle.load(f)
with open("/content/sample_data/unmon_standard10_3000.pkl", 'rb') as f:
    unmonitored_data = pickle.load(f)

# Assign labels
# Monitored data is labeled as 1
X_monitored = list(monitored_data.values())
y_monitored = [1] * len(X_monitored)

# Unmonitored data is labeled as -1
X_unmonitored = list(unmonitored_data)
y_unmonitored = [-1] * len(unmonitored_data)

# Combine the data and labels
X = X_monitored + X_unmonitored
y = y_monitored + y_unmonitored

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Training set size: {len(X_train)}")
print(f"Test set size: {len(X_test)}")
# first 5 data print
print("X_train (first 5):", X_train[:5])
print("y_train (first 5):", y_train[:5])

# first 5 data print
print("X_test (first 5):", X_test[:5])
print("y_test (first 5):", y_test[:5])



```
