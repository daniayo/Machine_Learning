```python
# Required Libraries for DBSCAN
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN

# Generate the dataset
x, y = make_moons(n_samples=300, noise=0.05, random_state=42)
df = pd.DataFrame(x)

# DBSCAN Clustering
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(x)

# Plot the DBSCAN Result
plt.figure(figsize=(7, 5))
plt.title("After - DBSCAN", fontsize=15)
plt.scatter(x[:, 0], x[:, 1], c=dbscan_labels)
plt.grid()
plt.show()
```

```python
# Required Libraries for KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans

# Generate the dataset
x, y = make_moons(n_samples=300, noise=0.05, random_state=42)
df = pd.DataFrame(x)

# KMeans Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(x)

# Plot the KMeans Result
plt.figure(figsize=(7, 5))
plt.title("After - kMeans", fontsize=15)
plt.scatter(x[:, 0], x[:, 1], c=kmeans_labels)
plt.grid()
plt.show()
```

- DBSCAN is density-based and excels in capturing irregular cluster shapes and handling noise, making it suitable for non-linear patterns.
- KMeans, based on Euclidean distance, assumes spherical clusters and fails to capture complex geometries, but it performs well for compact and convex datasets.
